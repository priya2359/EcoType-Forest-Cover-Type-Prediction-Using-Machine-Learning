## Target Variable Analysis

```python
ct_counts = df['Cover_Type'].value_counts().sort_index()
print("Cover_Type distribution:\n", ct_counts)
print(f"\nClass balance ratio (max/min): {ct_counts.max() / ct_counts.min():.2f}x")

fig, ax = plt.subplots(figsize=(10, 4))
ct_counts.plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
ax.set_title('Cover Type Class Distribution')
ax.set_xlabel('Cover Type')
ax.set_ylabel('Count')
ax.tick_params(axis='x', rotation=30)
plt.tight_layout()
plt.show()
```

## Categorical Features

```python
print(f"Wilderness_Area — {df['Wilderness_Area'].nunique()} unique values: {sorted(df['Wilderness_Area'].unique())}")
print(df['Wilderness_Area'].value_counts().sort_index().to_string())

print(f"\nSoil_Type — {df['Soil_Type'].nunique()} unique values: {sorted(df['Soil_Type'].unique())}")
print(df['Soil_Type'].value_counts().sort_index().to_string())
```

## Range & Validity Checks

```python
num_cols = df.select_dtypes(include='int64').columns
print("Feature value ranges:")
for col in num_cols:
    print(f"  {col:<45} [{df[col].min():>6}, {df[col].max():>6}]")

# Columns that should never be negative
non_neg_cols = [
    'Horizontal_Distance_To_Hydrology', 'Slope',
    'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
    'Horizontal_Distance_To_Roadways', 'Horizontal_Distance_To_Fire_Points'
]
print("\nNegative-value audit:")
for col in non_neg_cols:
    n = (df[col] < 0).sum()
    status = "CLEAN" if n == 0 else f"[!!] {n} negatives"
    print(f"  {col:<45} {status}")

print(f"\n  {'Vertical_Distance_To_Hydrology':<45} {(df['Vertical_Distance_To_Hydrology'] < 0).sum()} negatives (valid - terrain can sit below water source)")
print(f"\n  Aspect==360 (duplicate of 0 deg): {(df['Aspect'] == 360).sum()} rows -- will recode to 0 in cleaning")
```
